1. Extraer la información del archivo

```python
df_vacunas = pd.read_csv("drive/MyDrive/reto_pd_np/country_vaccinations.csv", index_col=0)
df_vacunas
```

2.  Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurandote que las columnas con fechas sean del tipo datetime64.

```python
df_vacunas.info()
```

```python
# Convertir date a tipo datatime64
df_vacunas["date"] = pd.to_datetime(df_vacunas["date"])
```

```python
# DataFrame actializado 
df_vacunas.info()
```

3. Determinar la cantidad de vacunas aplicadas de cada compañía (en base a como lo reporta cada país en la columna vaccines)

```python
vacunas_compania = df_vacunas.groupby("vaccines")["total_vaccinations"].sum()

df_vacunas_compania = pd.DataFrame({
    "Compañías": vacunas_compania.index,
    "Cantidad de vacunas aplicadas": vacunas_compania.values
})
df_vacunas_compania
```

4. Obtener la cantidad de vacunas aplicadas en todo el mundo



```python
total_vacunas = df_vacunas["total_vaccinations"].sum()
print("El total de vacunas aplicadas en el mundo es de: ",total_vacunas)
```

5. Calcular el promedio de vacunas aplicadas por país

```python
promedio_vacunas = df_vacunas.groupby("country")["total_vaccinations"].mean()
promedio_vacunas

df_promedio_vacunas = pd.DataFrame({
    "País": promedio_vacunas.index,
    "Promedio de vacunas aplicadas": promedio_vacunas.values
})
df_promedio_vacunas
```

6. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo

```python
vacunas_apli_dia = df_vacunas.loc[df_vacunas["date"] == "2021-01-29", "total_vaccinations"].sum()
print("El total de vacunas aplicadas el día 29/01/21 en todo el mundo es de: ", vacunas_apli_dia)
```

7. Crear un dataframe nuevo denoninado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_ra

```python
#Calculando las diferencias entre las columnas "daily_vaccinations" y "daily_vacinations_raw"
diferencias = df_vacunas["daily_vaccinations"] - df_vacunas["daily_vaccinations_raw"]

#Nuevo DataFrame con las columnas originales y la "diferencias"
conDiferencias = df_vacunas.loc[:, ["daily_vaccinations", "daily_vaccinations_raw"]].assign(diferencias=diferencias)
conDiferencias
```

8. Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua

```python
#obteniendo la fecha mas reciente y la más antigua
fecha_reciente = df_vacunas["date"].min()
fecha_antigua = df_vacunas["date"].max()

#Calculando diferencia en días
diferencia = (fecha_antigua - fecha_reciente).days

print(f'El período de tiempo entre el registro más antiguo y el más reciente es de {diferencia} días')
```

9. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la 
columna vaccines y separar por el carácter / )

```python
# creando columna llamada "canVac"
conCantidad = df_vacunas["canVac"] = df_vacunas['vaccines'].str.split('/').apply(lambda x: len(x))

df_vacunas[[ "date", "vaccines", "canVac"]]
```

10. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20 = df_vacunas[df_vacunas["date"] < "2020-12-20"]
antes20
```

11. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer

```python
pfizer = df_vacunas[df_vacunas["vaccines"].str.contains("Pfizer")]
pfizer
```

12. Almacenar los dataframes generados (conDiferencias,conCantidad,antes20 y pfizer) en 
un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una 
hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter

```python
writer = pd.ExcelWriter('resultadosReto.xlsx', engine='xlsxwriter')

# Guardar cada dataframe en una hoja del archivo de Excel
conDiferencias.to_excel(writer, sheet_name='conDiferencias')
conCantidad.to_excel(writer, sheet_name='conCantidad')
antes20.to_excel(writer, sheet_name='antes20')
pfizer.to_excel(writer, sheet_name='pfizer')

# Guardar y cerrar el archivo de Excel
writer.save()
```
